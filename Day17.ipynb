# Logistic Regression Using Sk Learn

```python
import pandas as pd 
```

```python
df = pd.read_csv("Sample DataSet/gym_logistic_regression_data.csv")
```

```python
df.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
le_fitness = LabelEncoder()
le_Diet = LabelEncoder()
le_gender = LabelEncoder()
df["Fitness_Level"] = le_fitness.fit_transform(df["Fitness_Level"])
df["Diet"] = le_fitness.fit_transform(df["Diet"])
df["Gender"] = le_fitness.fit_transform(df["Gender"])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
X = df.drop("Will_Subscribe", axis = 1)
y = df["Will_Subscribe"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import  accuracy_score, confusion_matrix
accuracy = accuracy_score(y_predict,y_test)
```
