# Machine Learning

Machine Learning is the ability of a system to learn from data without being explicitly programmed.

Examples:

YouTube recommendation

Face unlock

Spam filter

Basic idea

# Types of Machine Learning

Supervised Learning	 -> Learn from labeled data	              -> Email spam detection

Unsupervised Learning  -> Learn from patterns in unlabeled data -> Customer segmentation

Reinforcement Learning -> Learn by rewards & punishments        -> Self-driving cars, Games
